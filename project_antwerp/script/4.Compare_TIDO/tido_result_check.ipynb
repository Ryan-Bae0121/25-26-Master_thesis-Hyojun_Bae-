```python

import pandas as pd, numpy as np

pred_df = pd.read_csv('/project_antwerp/hbae/data/tido_prediction_results.csv')
gt_obs  = np.load('/project_antwerp/hbae/data/0228_HVG_NEW/combined_obs.npy', allow_pickle=True)
gt_obs_clean = set(o.replace('_hires','') for o in gt_obs)

def extract_spot_key(tile_id):
    if tile_id.startswith('Queensland P5'):
        parts = tile_id.split('_')
        return f'P5_{parts[2]}'
    parts = tile_id.split('_')
    if parts[0].startswith('GSE'):
        parts = parts[1:]
    elif parts[0] in ('Zenodo', 'Queensland'):
        parts = parts[1:]
    sample = parts[0]
    if sample.startswith('GSM6339'):
        if len(parts) > 2 and parts[1].startswith('s') and parts[1][1:].isdigit():
            return f'{sample}_{parts[1]}_{parts[2]}'
    if sample == 'Visium':
        return f'Visium_{parts[1]}_{parts[2]}' if len(parts) > 2 else f'Visium_{parts[1]}'
    barcode = parts[1] if len(parts) > 1 else ''
    return f'{sample}_{barcode}'

# unmatched 샘플별 카운트
from collections import Counter
unmatched_by_sample = Counter()
unmatched_examples = {}
for tid in pred_df['tile_id']:
    key = extract_spot_key(tid)
    if key not in gt_obs_clean:
        sample = '_'.join(tid.split('_')[:2])
        unmatched_by_sample[sample] += 1
        if sample not in unmatched_examples:
            unmatched_examples[sample] = (tid, key)

print('=== unmatched count by sample ===')
for s, cnt in unmatched_by_sample.most_common():
    tid, key = unmatched_examples[s]
    print(f'  {s}: {cnt}')
    print(f'    tile_id: {repr(tid)}')
    print(f'    key:     {repr(key)}')
```
